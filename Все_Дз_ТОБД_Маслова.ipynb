1.6 Написать фукнцию, которая для целых значений n и m будет возвращать матрицу размера n на m, имеющую заполнение вида.
Пример для 4 на 4:

0 1 0 1

2 3 2 3

0 1 0 1

2 3 2 3

````python
def n_m_marix(n,m):
    x = np.array([[1,0],[2,3]])
    numRow = int(n/2)+1
    numCol = int(m/2)+1
    arr=np.tile(x,(numRow,numCol))
    return arr[:n,:m]

n_m_marix(3,3)
````

1.7 Найти все простые числа в пределах ста. (Для решения предлагается использовать Решето Эратосфена) Использовать не более 1 цикла (желательно).

Для нахождения всех простых чисел не больше заданного числа n, следуя методу Эратосфена, нужно выполнить следующие шаги:

1. Выписать подряд все целые числа от двух до $n (2, 3, 4, …, n)$.
2. Пусть переменная $p$ изначально равна двум — первому простому числу.
3. Зачеркнуть в списке числа от $2p$ до $n$ считая шагами по $p$ (это будут числа кратные $p: 2p, 3p, 4p, …$).
4. Найти первое незачёркнутое число в списке, большее чем p, и присвоить значению переменной p это число.
5. Повторять шаги 3 и 4, пока возможно.
6. Теперь все незачёркнутые числа в списке — это все простые числа от 2 до $n$.

На практике, алгоритм можно улучшить следующим образом. На шаге №3 числа можно зачеркивать начиная сразу с числа $p^2$, потому что все составные числа меньше него уже будут зачеркнуты к этому времени. И, соответственно, останавливать алгоритм можно, когда $p^2$ станет больше, чем $n$. Также, все простые числа (кроме 2) — нечётные числа, и поэтому для них можно считать шагами по $2p$, начиная с $p^2$.

````python
a=np.arange(0,101)
a[1]=0
b=[]
for i in range(100):
    if a[i]!=0:
        b.append(a[i])
        j=i*2
        while j<=100:
            a[j]=0
            j+=i
b
````

````python
a = np.arange(2, 101, 1)
for i in range( len(a)):
    np.place(a, a%a[i]==0, a[i])
    d = np.delete(np.where(a==a[i])[0],0)
    if d.size==0:
        print(a)
        break
    a= np.delete(a,d)
````

    2.1.1 С применением рекурсии

````python
def recur_fibo(n):
    if n < 2:
        return n
    return(recur_fibo(n-1) + recur_fibo(n-2))
````

````python
%timeit recur_fibo(35)
````

    2.1.2 С применением рекурсии мемоизации
    
См.: 
* https://ru.wikipedia.org/wiki/Мемоизация
* https://docs.python.org/3/library/functools.html декоратор `lru_cache`
    

````python
from functools import lru_cache
````

````python
@lru_cache(maxsize=300)
def mem_fibo(n):
    if n < 2:
        return n
    return mem_fibo(n-1) + mem_fibo(n-2)
````

````python
%timeit mem_fibo(35)
````

    2.1.3 С применением циклов

````python
def cyc_fibo(n, suma = False):
    fibonacciSeries = [0,1]
    if n>2:
        for i in range(2, n):
            nextElement = fibonacciSeries[i-1] + fibonacciSeries[i-2]
            fibonacciSeries.append(nextElement)
    if suma:
        return sum(fibonacciSeries)
    else:
        return sum(fibonacciSeries[-2:])
````

При реализации использовать только цикл `for`

````python
from math import pi, sqrt
````

````python
def pi_N(N):
    a =[]
    for i in range(1,N+1):
        a.append(1/i**2)
    return sqrt(sum(a)*6)

pi_N(10000)
````

````python
@njit

def jit_pi_N1(arr):
    for i in range(1,len(arr)+1):
        arr[i-1] = 1/i**2
    return sqrt(arr.sum()*6)
````

````python
def countRussianLettersOccurrencesWithOutput(file, output):
    with open(mypath + file, "r", encoding="CP1251") as f:
        data = f.read().replace('\n', '').lower()

    counts = Counter(data)
    
    output.put(counts)
````

При помощи очереди передать задачи процессам и при помощи другой очереди забрать от них ответы. Определить за каое время решается задача для всех файлов из папки 'text'.

````python
import multiprocessing as mp
import queue # для перехвата иключения Empty, которое в модуле mp используется из queue
output = mp.Queue() # очередь для ответов
````

````python
tasks = mp.Queue()
[tasks.put(file) for file in files] # заполняем очередь с заданиями
````

````python
def countLettersWithFixedProccess(tasks, output):
    """  В блоке try бесконечный цикл, который идет в Tasks за очередным файлом для подсчета.
    Результат для каждого файла отправляется в queue.
    Когда очередь tasks оказывается пустой, на каждом потоке вылетает исключение Empty,
    которое просто заканчивает выполнение функции, т.к. вся работа выполнена, а результаты хранятся снаружи. 
    """
    
    try: 
        while True:
            file = tasks.get(False)
            with open(mypath + file, "r", encoding="CP1251") as f:
                data = f.read().replace('\n', '').lower()
    
            counts = Counter(data)

            output.put(counts)            

    except queue.Empty:
        pass
````

````python
def sumResultsWithFixedProccess():
    start_time = time.time()
    sumResult = Counter()
    processes = [mp.Process(target=countLettersWithFixedProccess, args=(tasks, output)) for x in range(mp.cpu_count())]
    
    for p in processes:
        p.start()

    for p in processes:
        p.join()

    results = [output.get() for p in files]
    
    for result in results:
        sumResult = sumResult+ result
    time_4_w_mp_fixed = time.time() - start_time
    print(f"Затраченное в секундах время: {time_4_w_mp_fixed}")

    countWrapper(sumResult)
````

Подсчитать частоту (!) слов для отзывов с оценкой "5" и для отзывов с оценкой "2" и найти 50 распространенных слов, частота которых максимально увеличивается и уменьшается при переходе от плохих отзывов к хорошим (ориентруйтесь на цель задачи, а не на форму).

````python
import dask.bag as db

filename = 'reviews_Toys_and_Games_5.json.gz'
lines = db.read_text(filename)

import json

js = lines.map(json.loads)
````

````python
import nltk
from nltk.corpus import stopwords

# useless words for lor task
useless = ['play','get','game','one','toy','make','year','time','also','piece','put','take','old','really','would','son','daughter','go','come','child','look','think','color','card','two','puzzle','still','say','together','dont']

stops = stopwords.words('english')+['']+list('1234567890')+useless

import nltk
from nltk.stem.wordnet import WordNetLemmatizer
lemma = WordNetLemmatizer()

import string


def count_text(d):
#     splits original text into words without punctuation
    words_orig = list(d['reviewText'].lower()
                                    .translate(str.maketrans('', '', string.punctuation))
                                    .split(' '))

#     removing 3rd form from verbs and plural forms of nouns
    tokens = [lemma.lemmatize(word, pos = "v") for word in words_orig] # verbs
    tokens = [lemma.lemmatize(word, pos = "n") for word in tokens] # nouns

#     new list for words
    words = []
    
#     removing stop words 
    for word in tokens:
        if word not in stops:
            words.append(word)
      

    return {'text': words}
````

````python
# function to flatten [[]] into []
flatten = lambda t: [item for sublist in t for item in sublist]
````

````python
from collections import Counter
````

````python
%%time

# big list of all words of rating '5'
all_words_5 = (flatten(js.filter(lambda x: x['overall']==5.0)
 .map(count_text)
 .pluck('text')
 .compute()))

# big list of all words of rating '2'
all_words_2 = (flatten(js.filter(lambda x: x['overall']==2.0)
 .map(count_text)
 .pluck('text')
 .compute()))

# counts word - quantity of ratings 5&2
count_5 = dict(Counter(all_words_5).most_common(len(all_words_5)))
count_2 = dict(Counter(all_words_2).most_common(len(all_words_2)))
````

````python
import operator
````

````python
# functions to write + or - before second element
plus = lambda list_x: [[i[0],'+ '+str(i[1])] for i in list_x]
minus = lambda list_x: [[i[0],'- '+str(i[1])] for i in list_x]
````

````python
# for 5

good_increase = []
good_decrease = []

for word in count_5:
#     test wether word is in count_2
    if word in count_2:
#         if more in 5
        if count_5[word] > count_2[word]:
#         add to list 'increase in good ones'
            good_increase.append([word, abs(count_2[word] - count_5[word])])

#         if less in 5
        elif count_5[word] < count_2[word]:
#         add to list 'decrease in good ones'
            good_decrease.append([word, abs(count_2[word] - count_5[word])])

#   the word is only in count_5
    else:
#         let's make it 100 as a threshold
        if count_5[word] > 100:
#             print(word)
            good_increase.append([word, count_5[word]])
  
# sorting
good_increase = sorted(good_increase, key=operator.itemgetter(1), reverse=True)
good_decrease = sorted(good_decrease, key=operator.itemgetter(1), reverse=True)

# add + & -
good_increase = plus(good_increase)
good_decrease = minus(good_decrease)
````

````python
# for 2

bad_increase = []
bad_decrease = []

for word in count_2:
#     test wether word is in count_5
    if word in count_5:
#         if more in 2
        if count_2[word] > count_5[word]:
#         add to list 'increase in bad ones'
            bad_increase.append([word, abs(count_2[word] - count_5[word])])

#         if less in 2
        elif count_2[word] < count_5[word]:
#         add to list 'decrease in bad ones'
            bad_decrease.append([word, abs(count_2[word] - count_5[word])])
    
#   the word is only in count_2
    else:
#         let's make it 10 as a threshold
        if count_2[word] > 5:
#             print(word)
            bad_increase.append([word, count_2[word]])
    

# sorting
bad_increase = sorted(bad_increase, key=operator.itemgetter(1), reverse=True)
bad_decrease = sorted(bad_decrease, key=operator.itemgetter(1), reverse=True)

# add + & -
bad_increase = plus(bad_increase)
bad_decrease = minus(bad_decrease)
````
